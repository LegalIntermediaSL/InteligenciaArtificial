## Calculadora de ROI para RPA con Computer Use

```python
from dataclasses import dataclass

@dataclass
class CalculadoraROI:
    nombre_proceso: str
    tiempo_manual_min: float
    tareas_por_dia: int
    coste_hora_empleado: float
    coste_por_ejecucion_api: float
    dias_laborables_mes: int = 22

    @property
    def coste_manual_diario(self) -> float:
        return (self.tiempo_manual_min / 60) * self.tareas_por_dia * self.coste_hora_empleado

    @property
    def coste_automatizado_diario(self) -> float:
        return self.tareas_por_dia * self.coste_por_ejecucion_api

    @property
    def ahorro_diario(self) -> float:
        return self.coste_manual_diario - self.coste_automatizado_diario

    @property
    def roi_mensual(self) -> float:
        return self.ahorro_diario * self.dias_laborables_mes

    def imprimir(self):
        print(f'\n📊 {self.nombre_proceso}')
        print(f'  Coste manual diario:       €{self.coste_manual_diario:.0f}')
        print(f'  Coste automatizado diario: €{self.coste_automatizado_diario:.0f}')
        print(f'  Ahorro diario:             €{self.ahorro_diario:.0f}')
        print(f'  ROI mensual:               €{self.roi_mensual:.0f}')
        pct = (self.ahorro_diario / self.coste_manual_diario) * 100 if self.coste_manual_diario > 0 else 0
        print(f'  Reducción de coste:        {pct:.0f}%')

casos = [
    CalculadoraROI('Introducción de facturas en ERP', tiempo_manual_min=15, tareas_por_dia=80,
                   coste_hora_empleado=25, coste_por_ejecucion_api=0.15),
    CalculadoraROI('Triaje de emails de soporte', tiempo_manual_min=3, tareas_por_dia=200,
                   coste_hora_empleado=20, coste_por_ejecucion_api=0.05),
    CalculadoraROI('Actualización de CRM', tiempo_manual_min=10, tareas_por_dia=50,
                   coste_hora_empleado=30, coste_por_ejecucion_api=0.20),
]

for caso in casos:
    caso.imprimir()

total_roi = sum(c.roi_mensual for c in casos)
print(f'\n💰 ROI total mensual (todos los procesos): €{total_roi:.0f}')
```

## Plantilla de agente para casos empresariales

```python
import anthropic
client = anthropic.Anthropic()

class AgenteEmpresarial:
    """Plantilla reutilizable para agentes Computer Use empresariales."""

    def __init__(self, nombre: str, sistema: str, url_objetivo: str):
        self.nombre = nombre
        self.sistema = sistema
        self.url_objetivo = url_objetivo
        self.historial_tareas = []

    def generar_instruccion(self, tarea: dict) -> str:
        """Genera instrucción específica a partir de datos de negocio."""
        campos = '\n'.join(f'  - {k}: {v}' for k, v in tarea.items())
        return f"""En {self.url_objetivo}:\n{campos}\n\nCompleta la tarea y confirma el resultado."""

    def ejecutar(self, tarea: dict) -> dict:
        instruccion = self.generar_instruccion(tarea)
        print(f'[{self.nombre}] Ejecutando: {list(tarea.keys())}')
        # En producción: llamar al bucle de Computer Use
        resultado = {'estado': 'simulado', 'instruccion_generada': instruccion[:100]}
        self.historial_tareas.append({'tarea': tarea, 'resultado': resultado})
        return resultado

# Ejemplo: agente de facturas
agente_facturas = AgenteEmpresarial(
    nombre='AgenteFacturas',
    sistema='Introduce facturas en el ERP con precisión contable.',
    url_objetivo='https://erp.empresa.com/facturas/nueva',
)

lote_facturas = [
    {'proveedor': 'TechSupplies SL', 'fecha': '2025-04-21', 'importe': 1250.00, 'concepto': 'Hardware'},
    {'proveedor': 'CloudServices SA', 'fecha': '2025-04-21', 'importe': 399.99, 'concepto': 'Suscripción cloud'},
]

for factura in lote_facturas:
    resultado = agente_facturas.ejecutar(factura)
    print(f'  → Estado: {resultado["estado"]}')

print(f'\nTotal procesadas: {len(agente_facturas.historial_tareas)}')
```
