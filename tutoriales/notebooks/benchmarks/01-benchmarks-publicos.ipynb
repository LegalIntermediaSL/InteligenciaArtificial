```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 1. Tabla comparativa de benchmarks

```python
datos = {
    'Modelo': ['Claude Opus 4.7', 'Claude Sonnet 4.6', 'Claude Haiku 4.5', 'GPT-4o', 'Gemini 1.5 Pro', 'Llama 3.1 70B'],
    'MMLU (%)': [90, 85, 78, 88, 86, 82],
    'HumanEval (%)': [92, 88, 75, 90, 85, 81],
    'GSM8K (%)': [97, 93, 82, 95, 91, 88],
    'Input $/M': [15.0, 3.0, 0.80, 2.50, 3.50, 0.50],
}

df = pd.DataFrame(datos).set_index('Modelo')
print('Benchmarks y precios por modelo:')
print(df.to_string())
```

## 2. Visualización de benchmarks

```python
benchmarks = ['MMLU (%)', 'HumanEval (%)', 'GSM8K (%)']
modelos = df.index.tolist()
x = np.arange(len(benchmarks))
width = 0.12

fig, ax = plt.subplots(figsize=(12, 6))
colores = ['#2196F3', '#4CAF50', '#FF9800', '#9C27B0', '#F44336', '#00BCD4']

for i, (modelo, color) in enumerate(zip(modelos, colores)):
    valores = [df.loc[modelo, b] for b in benchmarks]
    ax.bar(x + i * width, valores, width, label=modelo, color=color, alpha=0.85)

ax.set_xlabel('Benchmark')
ax.set_ylabel('Score (%)')
ax.set_title('Comparativa de benchmarks públicos de LLMs (2025)')
ax.set_xticks(x + width * (len(modelos) - 1) / 2)
ax.set_xticklabels([b.replace(' (%)', '') for b in benchmarks])
ax.set_ylim(70, 100)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3. Ratio calidad/precio

```python
df['Score medio'] = df[['MMLU (%)', 'HumanEval (%)', 'GSM8K (%)']].mean(axis=1)
df['Score/Dolar'] = df['Score medio'] / df['Input $/M']

print('\nRatio calidad/precio (score medio / precio input):')
ranking = df[['Score medio', 'Input $/M', 'Score/Dolar']].sort_values('Score/Dolar', ascending=False)
print(ranking.round(1).to_string())

print('\n💡 Mayor Score/Dolar = mejor relación calidad/precio')
```

## Conclusión
