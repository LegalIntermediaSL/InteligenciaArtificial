## 1. La jerarquía IA → ML → DL → LLMs

Visualizamos la relación de inclusión entre los conceptos principales.

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title('La jerarquía de la IA', fontsize=16, fontweight='bold', pad=20)

# Círculos concéntricos
circulos = [
    (5, 5, 4.5, '#AED6F1', 'Inteligencia Artificial (IA)', 1.0, 14),
    (5, 5, 3.2, '#A9DFBF', 'Machine Learning (ML)', 2.0, 13),
    (5, 5, 2.0, '#F9E79F', 'Deep Learning (DL)', 3.3, 12),
    (5, 5, 1.0, '#F1948A', 'LLMs / Transformers', 4.2, 10),
]

for cx, cy, radio, color, etiqueta, y_text, fontsize in circulos:
    circulo = plt.Circle((cx, cy), radio, color=color, alpha=0.6)
    ax.add_patch(circulo)
    ax.text(cx, y_text, etiqueta, ha='center', va='center',
            fontsize=fontsize, fontweight='bold')

plt.tight_layout()
plt.show()
print('IA es el campo más amplio. LLMs son la tecnología más específica y reciente.')
```

## 2. Crecimiento de los modelos de lenguaje

Visualizamos la evolución exponencial del número de parámetros.

```python
modelos = {
    'GPT-1 (2018)': 0.117,
    'GPT-2 (2019)': 1.5,
    'GPT-3 (2020)': 175,
    'PaLM (2022)': 540,
    'GPT-4 (2023)*': 1000,
    'Llama 3 405B (2024)': 405,
}

nombres = list(modelos.keys())
params = list(modelos.values())
colores = ['#3498DB', '#2ECC71', '#E74C3C', '#9B59B6', '#F39C12', '#1ABC9C']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Escala lineal
bars = ax1.barh(nombres, params, color=colores)
ax1.set_xlabel('Miles de millones de parámetros')
ax1.set_title('Parámetros por modelo (escala lineal)')
ax1.bar_label(bars, fmt='%.1fB', padding=3)

# Escala logarítmica
ax2.barh(nombres, params, color=colores)
ax2.set_xscale('log')
ax2.set_xlabel('Miles de millones de parámetros (escala log)')
ax2.set_title('Parámetros por modelo (escala logarítmica)')

plt.suptitle('Evolución del tamaño de los LLMs', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
print('* Estimación no oficial')
```

## 3. Hablando con un LLM desde Python

Ejemplo práctico: preguntar a Claude sobre conceptos de IA.

```python
# Configura tu API key antes de ejecutar esta celda
# Opción 1: variable de entorno ANTHROPIC_API_KEY
# Opción 2: descomentar la línea siguiente y pegar tu clave
# os.environ['ANTHROPIC_API_KEY'] = 'sk-ant-...'

import anthropic

client = anthropic.Anthropic()

def preguntar_a_claude(pregunta: str, max_tokens: int = 300) -> str:
    """Hace una pregunta a Claude y devuelve la respuesta."""
    respuesta = client.messages.create(
        model='claude-haiku-4-5-20251001',
        max_tokens=max_tokens,
        messages=[{'role': 'user', 'content': pregunta}]
    )
    return respuesta.content[0].text

# Prueba básica
respuesta = preguntar_a_claude(
    'En una sola frase, ¿cuál es la diferencia entre Machine Learning y Deep Learning?'
)
print('Claude responde:')
print(respuesta)
```

```python
# Explorar capacidades: pedir ejemplos de aplicaciones de IA por sector
sectores = ['salud', 'finanzas', 'educación']

for sector in sectores:
    respuesta = preguntar_a_claude(
        f'Da un ejemplo concreto y real de aplicación de IA en el sector de {sector}. '
        f'Una frase máximo.',
        max_tokens=100
    )
    print(f'\n🏥 {sector.capitalize()}: {respuesta}')
```
