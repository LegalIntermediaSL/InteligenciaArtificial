```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

## 1. Timeline de hitos clave

```python
hitos = [
    (1950, 'Test de Turing', 'fundamentos'),
    (1956, 'Conferencia Dartmouth\n(nace el término IA)', 'fundamentos'),
    (1957, 'Perceptrón\n(Rosenblatt)', 'ml'),
    (1969, '1er Invierno de la IA\n(DARPA recorta fondos)', 'invierno'),
    (1980, 'Sistemas expertos\n(auge comercial)', 'fundamentos'),
    (1987, '2do Invierno de la IA', 'invierno'),
    (1997, 'Deep Blue vence\na Kasparov (ajedrez)', 'ml'),
    (2006, 'Deep Learning\n(Hinton, preentrenamiento)', 'dl'),
    (2012, 'AlexNet gana ImageNet\n(inicio era Deep Learning)', 'dl'),
    (2017, 'Transformers\n("Attention is All You Need")', 'llm'),
    (2018, 'BERT / GPT-1\n(LLMs preentrenados)', 'llm'),
    (2020, 'GPT-3\n(175B parámetros)', 'llm'),
    (2022, 'ChatGPT\n(IA accesible para todos)', 'llm'),
    (2023, 'Claude, Llama, Gemini\n(ecosistema diverso)', 'llm'),
    (2025, 'Claude 4.X, Agentes\nCommunity Use, AGI en debate', 'llm'),
]

colores_tipo = {
    'fundamentos': '#2196F3',
    'ml': '#4CAF50',
    'dl': '#FF9800',
    'llm': '#9C27B0',
    'invierno': '#F44336',
}

fig, ax = plt.subplots(figsize=(16, 6))

# Línea de tiempo
ax.axhline(y=0, color='gray', linewidth=2, zorder=1)

for i, (year, evento, tipo) in enumerate(hitos):
    y = 0.4 if i % 2 == 0 else -0.4
    color = colores_tipo[tipo]
    ax.scatter(year, 0, s=80, color=color, zorder=3)
    ax.annotate(
        f'{year}\n{evento}',
        xy=(year, 0), xytext=(year, y),
        ha='center', va='bottom' if y > 0 else 'top',
        fontsize=7, color=color, fontweight='bold',
        arrowprops=dict(arrowstyle='-', color=color, lw=0.8),
    )

# Leyenda
leyenda = [mpatches.Patch(color=c, label=t.upper()) for t, c in colores_tipo.items()]
ax.legend(handles=leyenda, loc='upper left', fontsize=8)

ax.set_xlim(1945, 2027)
ax.set_ylim(-1, 1)
ax.set_title('Historia de la Inteligencia Artificial (1950-2025)', fontsize=14, fontweight='bold')
ax.set_xlabel('Año')
ax.get_yaxis().set_visible(False)
for spine in ['left', 'top', 'right']:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```

## 2. Crecimiento de parámetros en LLMs

```python
modelos_params = [
    ('GPT-1 (2018)', 117e6),
    ('BERT (2018)', 340e6),
    ('GPT-2 (2019)', 1.5e9),
    ('GPT-3 (2020)', 175e9),
    ('PaLM (2022)', 540e9),
    ('GPT-4 (2023)', 1.8e12),
    ('Claude 3 Opus (2024)', 2e12),
]

nombres = [m[0] for m in modelos_params]
params = [m[1] for m in modelos_params]

fig, ax = plt.subplots(figsize=(12, 5))
colores = plt.cm.viridis(np.linspace(0.2, 0.9, len(nombres)))
bars = ax.barh(nombres, params, color=colores)
ax.set_xscale('log')
ax.set_xlabel('Número de parámetros (escala logarítmica)')
ax.set_title('Crecimiento de parámetros en LLMs', fontsize=13, fontweight='bold')

for bar, param in zip(bars, params):
    if param >= 1e12:
        label = f'{param/1e12:.1f}T'
    elif param >= 1e9:
        label = f'{param/1e9:.0f}B'
    else:
        label = f'{param/1e6:.0f}M'
    ax.text(bar.get_width() * 1.1, bar.get_y() + bar.get_height()/2, label, va='center', fontsize=9)

plt.tight_layout()
plt.show()
print('\n⚡ De 117M a 2T parámetros en 6 años: ×17.000 de crecimiento')
```

## 3. Los dos "inviernos" de la IA y por qué no volverán (fácilmente)

```python
razones_inviernos = {
    '1er Invierno (1969-1980)': [
        'El Perceptrón no podía resolver XOR (Minsky & Papert)',
        'Informe DARPA limitó financiación',
        'Expectativas no cumplidas',
    ],
    '2do Invierno (1987-1993)': [
        'Sistemas expertos frágiles y costosos',
        'Hardware insuficiente para redes neuronales',
        'LISP machines superadas por PCs',
    ],
    'Por qué es diferente hoy (2025)': [
        '✅ Aplicaciones comerciales reales (ChatGPT: 100M usuarios en 2 meses)',
        '✅ Hardware (GPUs/TPUs) sigue creciendo según ley de Moore adaptada',
        '✅ Datos: internet genera zettabytes anuales',
        '✅ Inversión privada: más de $200B en IA en 2023-2024',
        '✅ Infraestructura cloud democratiza el acceso',
    ],
}

for periodo, puntos in razones_inviernos.items():
    print(f'\n📌 {periodo}:')
    for punto in puntos:
        print(f'   • {punto}')
```
