## 1. Taxonomía de la IA

```
Inteligencia Artificial
├── Narrow AI (IA estrecha) — sistemas actuales
│   ├── Machine Learning
│   │   ├── Supervisado (clasificación, regresión)
│   │   ├── No supervisado (clustering, reducción de dim.)
│   │   └── Por refuerzo (juegos, robótica)
│   └── Deep Learning
│       ├── CNNs — visión por computador
│       ├── RNNs/LSTMs — series temporales
│       └── Transformers — LLMs, BERT, GPT
└── AGI (IA general) — hipotética, no existe aún
```

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(12, 7))
ax.set_xlim(0, 10)
ax.set_ylim(0, 7)
ax.axis('off')
ax.set_title('Jerarquía de tipos de IA', fontsize=16, fontweight='bold', pad=20)

# Cajas del diagrama
cajas = [
    (3.5, 6, 3, 0.6, '#E3F2FD', 'Inteligencia Artificial', 12),
    (1, 5, 3, 0.6, '#BBDEFB', 'Narrow AI (actual)', 11),
    (6, 5, 3, 0.6, '#E8F5E9', 'AGI (hipotética)', 11),
    (0.5, 3.8, 2, 0.6, '#C8E6C9', 'Machine Learning', 10),
    (2.7, 3.8, 2, 0.6, '#FFF9C4', 'Deep Learning', 10),
    (0, 2.6, 1.5, 0.6, '#F8BBD9', 'Supervisado', 9),
    (1.6, 2.6, 1.5, 0.6, '#F8BBD9', 'No supervisado', 9),
    (2.5, 2.6, 1.5, 0.6, '#FFE0B2', 'CNNs (visión)', 9),
    (4.1, 2.6, 1.5, 0.6, '#FFE0B2', 'Transformers', 9),
]

for x, y, w, h, color, texto, fontsize in cajas:
    rect = patches.FancyBboxPatch((x, y), w, h, boxstyle='round,pad=0.05', facecolor=color, edgecolor='gray', linewidth=1)
    ax.add_patch(rect)
    ax.text(x + w/2, y + h/2, texto, ha='center', va='center', fontsize=fontsize, fontweight='bold')

plt.tight_layout()
plt.show()
````

## 2. Clasificación con ML supervisado (ejemplo práctico)

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Narrow AI: clasificador supervisado
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

modelo = DecisionTreeClassifier(max_depth=3)
modelo.fit(X_train, y_train)

predicciones = modelo.predict(X_test)
accuracy = accuracy_score(y_test, predicciones)

print(f'Clasificador Iris (Narrow AI — ML Supervisado)')
print(f'Accuracy: {accuracy:.1%}')
print(f'\nEste modelo SOLO sabe clasificar flores Iris.')
print(f'Narrow AI = excelente en una tarea específica, inútil fuera de ella.')
````

## Resumen

- **Narrow AI**: sistemas actuales, excelentes en una tarea específica
- **Machine Learning**: aprende patrones de datos sin ser programado explícitamente
- **Deep Learning**: redes neuronales profundas, domina imágenes, audio y texto
- **LLMs**: Transformers entrenados en billones de tokens, razonamiento multi-dominio
- **AGI**: inteligencia general comparable a la humana — hipotética, no existe aún
